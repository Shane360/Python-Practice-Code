```python
table1 = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                      "gender" : ("male", "male", "female", "female", "female", "male", "female", "male"),
                      "height" : (71,73,64,66,69,62,72, 69),
                       "weight" : (175,225,130,125,165,160,180,165)})
table1
```

```python
table2 = pd.DataFrame({"P_ID": (1,2,4,5,7,8,9,10),
                      "sex": ("male", "male", "female", "female", "female", "male", "male", "female"),
                      "visits": (1,2,3,12,2,2,1,1),
                      "checkup": (1,1,1,1,1,1,0,0),
                      "follow_up": (0,0,1,2,0,0,0,0),
                       "illness": (0,0,2,7,1,1,0,0),
                       "surgery": (0,0,0,2,0,0,0,0),
                       "ER": (0,1,0,0,0,0,1,1) })

table2
```

```python
combined_table = pd.merge(table1,
                         table2,
                         how = "inner",
                         on = "P_ID")
combined_table
```

```python
left_join = pd.merge(table1,
                     table2,
                     how = "left",
                    on = "P_ID")
left_join
```

```python
outer_join = pd.merge(table1,
                     table2,
                     how = "outer",
                     on = "P_ID")
outer_join
```

```python
table1.rename(columns = {"gender":"sex"}, inplace = True)
combined2 = pd.merge(table1,
                    table2,
                    how = "outer",
                    on = ["P_ID", "sex"])
combined2
```
