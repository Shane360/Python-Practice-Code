```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
diamonds = pd.read_csv("/kaggle/input/diamonds/diamonds.csv")
print(diamonds.shape)

diamonds.head(15)
```

```python
diamonds.hist(column = "carat",
             figsize = (8,8),
             color = "teal");
```

```python
diamonds.hist(column = "carat",
             figsize = (8,8),
             color = "teal",
             bins = 50,
             range = (0, 3.5));
```

```python
diamonds.boxplot(column = "carat");
```

```python
diamonds.boxplot(column = "price",
                by = "clarity",
                 figsize = (8, 8));
```

```python
diamonds["carat"].plot(kind = "density",
                       figsize = (8, 8),
                       xlim = (0, 5));
```

```python
carat_table = pd.crosstab(index = diamonds["clarity"], columns = "count")
carat_table
```

```python
carat_table.plot(kind = "bar",
                        figsize = (8, 8));
```

```python
carat_table = pd.crosstab(index = diamonds["clarity"],
                         columns = diamonds["color"])
carat_table
```

```python
carat_table.plot(kind = "bar",
                figsize = (8, 8),
                stacked = True);
```

```python
carat_table.plot(kind = "bar",
                figsize = (8, 8),
                stacked = False);
```

```python
diamonds.plot(kind = "scatter",
             x = "carat",
             y = "price",
             figsize = (0,10),
             ylim = (0,2000));
```

```python
#Create some data
years = [y for y in range(1950, 2016)]
readings = [(y + np.random.uniform(0, 20) - 1900) for y in years]
time_df = pd.DataFrame({"year": years,
                       "readings": readings})

#Plot the data
time_df.plot(x = "year",
            y = "readings",
            figsize = (9, 9));
```

```python
my_plot = time_df.plot(x = "year",         #Create the plot and save to a variable
                      y = "readings",
                      figsize = (9, 9))
my_fig = my_plot.get_figure()               #Get the figure

my_fig.savefig("line_plot_example.png")     #Save to file.
```
